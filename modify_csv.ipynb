```python
import pandas as pd

# Define file paths
file_paths = [
    '/Users/ioannisveroulis/premier_league_predictions_over2.5/data/2020-2021.csv',
    '/Users/ioannisveroulis/premier_league_predictions_over2.5/data/2021-2022.csv',
    '/Users/ioannisveroulis/premier_league_predictions_over2.5/data/2022-2023.csv',
    '/Users/ioannisveroulis/premier_league_predictions_over2.5/data/2023-204.csv',
    '/Users/ioannisveroulis/premier_league_predictions_over2.5/data/2024-2025.csv'
]

# List to store DataFrames
df_list = []

# Loop through each file path, process the file, and add it to the list
for file_path in file_paths:
    # Load the CSV file
    df = pd.read_csv(file_path)
    
    # Add "over2.5" column
    df['over2.5'] = (df['FTHG'] + df['FTAG']) > 2.5
    df['over2.5'] = df['over2.5'].astype(int)
    
    # Add "goal-goal" column
    df['goal-goal'] = (df['FTHG'] > 0) & (df['FTAG'] > 0)
    df['goal-goal'] = df['goal-goal'].astype(int)
    
    # Rename the columns
    df.rename(columns={
        'FTHG': 'HomeResult',
        'FTAG': 'AwayResult',
        'FTR': 'Result',
        'HTHG': 'Home_HalfTime',
        'HTAG': 'Away_HalfTime',
        'HTR': 'HalfTimeResult'
    }, inplace=True)
    
    # Delete the "Referee" column if it exists
    if 'Referee' in df.columns:
        df.drop(columns=['Referee'], inplace=True)
    
    # Append the modified DataFrame to the list
    df_list.append(df)

# Concatenate all DataFrames in the list
combined_df = pd.concat(df_list, ignore_index=True)

# Save the concatenated DataFrame to a new CSV file
output_file_path = '/Users/ioannisveroulis/premier_league_predictions_over2.5/data/combined_with_over2.5_goal_goal.csv'
combined_df.to_csv(output_file_path, index=False)

print(f"Combined CSV file saved to {output_file_path}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report
import joblib

# Load the combined CSV file
combined_csv_path = '/Users/ioannisveroulis/premier_league_predictions_over2.5/data/combined_with_over2.5_goal_goal.csv'
df = pd.read_csv(combined_csv_path)

# Check if the target columns are present; if not, create them
if 'over2.5' not in df.columns:
    df['over2.5'] = (df['HomeResult'] + df['AwayResult']) > 2.5
    df['over2.5'] = df['over2.5'].astype(int)

if 'goal-goal' not in df.columns:
    df['goal-goal'] = (df['HomeResult'] > 0) & (df['AwayResult'] > 0)
    df['goal-goal'] = df['goal-goal'].astype(int)

# Select features (Bet365 odds) and target variables
features = ['B365>2.5', 'B365<2.5']
X = df[features]
y_over2_5 = df['over2.5']
y_goal_goal = df['goal-goal']

# Split the data into training and testing sets
X_train_over, X_test_over, y_train_over, y_test_over = train_test_split(X, y_over2_5, test_size=0.2, random_state=42)
X_train_goal, X_test_goal, y_train_goal, y_test_goal = train_test_split(X, y_goal_goal, test_size=0.2, random_state=42)

# Define the model and grid search parameters
param_grid = {
    'n_estimators': [100, 200, 500],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Train the model for over 2.5 goals prediction
rf_model_over = RandomForestClassifier(random_state=42)
grid_search_over = GridSearchCV(estimator=rf_model_over, param_grid=param_grid, cv=5, n_jobs=-1, scoring='roc_auc', verbose=2)
grid_search_over.fit(X_train_over, y_train_over)
best_rf_model_over = grid_search_over.best_estimator_

# Train the model for goal-goal prediction
rf_model_goal = RandomForestClassifier(random_state=42)
grid_search_goal = GridSearchCV(estimator=rf_model_goal, param_grid=param_grid, cv=5, n_jobs=-1, scoring='roc_auc', verbose=2)
grid_search_goal.fit(X_train_goal, y_train_goal)
best_rf_model_goal = grid_search_goal.best_estimator_

# Evaluate the models
y_pred_over = best_rf_model_over.predict(X_test_over)
y_pred_goal = best_rf_model_goal.predict(X_test_goal)

print(f"Over 2.5 Goals Prediction Accuracy: {accuracy_score(y_test_over, y_pred_over)}")
print(f"Goal-Goal Prediction Accuracy: {accuracy_score(y_test_goal, y_pred_goal)}")

# Save the models for future predictions
joblib.dump(best_rf_model_over, 'best_rf_model_over.pkl')
joblib.dump(best_rf_model_goal, 'best_rf_model_goal.pkl')

print("Models saved successfully.")

# Example prediction for a new match
def predict_new_match(over_model, goal_model, B365_over, B365_under):
    match_data = pd.DataFrame([[B365_over, B365_under]], columns=['B365>2.5', 'B365<2.5'])
    over_prediction = over_model.predict(match_data)[0]
    goal_prediction = goal_model.predict(match_data)[0]
    return over_prediction, goal_prediction

# Replace these odds with the odds of the new match you want to predict
B365_over = 1.85
B365_under = 1.95

over_pred, goal_pred = predict_new_match(best_rf_model_over, best_rf_model_goal, B365_over, B365_under)
print(f"Prediction for new match: Over 2.5 -> {over_pred}, Goal-Goal -> {goal_pred}")
```
